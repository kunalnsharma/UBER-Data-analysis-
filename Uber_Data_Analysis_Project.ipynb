```python
# Data Cleaning

#First we have to change the dataype of Date columns object to Datetime format

df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Now we have to fill the NaN values to 0 because in other cells values are fill means there are some reason and if there are no reason so it will be show as NaN so we have to change it

df = df.fillna(0)
```

```python
# Now we have to make a month columns to clear more about the data there are data only for year 2024

df['month'] = df['Date'].dt.month
```

```python
cancelled_df = df[df['Booking Status'].isin(['Cancelled', 'Incomplete', 'No Driver Found'])]

# Cancellation reasons count
reason_counts = cancelled_df['Driver Cancellation Reason'].value_counts()
print(reason_counts)
```

```python
df['Cancellation_Type'] = df['Booking Status'].apply(lambda x:
    'Driver' if x == 'Cancelled'
    else ('Customer' if x == 'Incomplete'
    else ('System' if x == 'No Driver Found'
    else 'Completed'))
)

print(df['Cancellation_Type'].value_counts())
```

```python

# Hourly cancellation count
hourly_cancellation = cancelled_df['hour'].value_counts().sort_index()
print(hourly_cancellation)

# Plot
sns.barplot(x=hourly_cancellation.index, y=hourly_cancellation.values)
plt.xlabel("Hour of Day")
plt.ylabel("Number of Cancellations")
plt.title("Cancellations by Hour")
plt.show()
```

```python
# The most cancelatiion of the booking is in between 4PM to 8PM
```

```python
pay_counts = df['Payment Method'].value_counts()
print(pay_counts)

sns.countplot(data=df, x='Payment Method', order=pay_counts.index)
plt.title("Payment Method Preference")
plt.show()
```

```python
# We can easily see here in the upper graph UPI is use in the most of booking around 47% of the time
```

```python
# Cancellation flag
df['is_cancelled'] = df['Booking Status'].isin(['Cancelled','No Driver Found','Incomplete']).astype(int)

cancel_ratio = df.groupby('Vehicle Type')['is_cancelled'].mean().sort_values(ascending=False) * 100
print(cancel_ratio)

cancel_ratio.plot(kind='bar')
plt.ylabel("Cancellation Ratio (%)")
plt.title("Cancellation Ratio by Vehicle Type")
plt.show()
```

```python
total_revenue = df['Booking Value'].sum()
avg_revenue = df['Booking Value'].mean()

print(f"Total Revenue: {total_revenue:.2f}")
print(f"Average Booking Value: {avg_revenue:.2f}")
```

```python
rev_by_type = df.groupby('Vehicle Type')['Booking Value'].sum().sort_values(ascending=False)
print(rev_by_type)

rev_by_type.plot(kind='bar')
plt.title("Revenue by Vehicle Type")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Conclusion :- as we see before the most used vehicle in the booking is AUTO so obvisouly we can see the revenue of the AUTO booking is higher then the other
```

```python
top_pickup = df['Pickup Location'].value_counts().head(10)
top_drop = df['Drop Location'].value_counts().head(10)

sns.barplot(y=top_pickup.index, x=top_pickup.values)
plt.title("Top 10 Pickup Locations")
plt.show()

sns.barplot(y=top_drop.index, x=top_drop.values)
plt.title("Top 10 Drop Locations")
plt.show()
```
